```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('21BCE3252 DATASET.csv')
```

```python
# Dimension and Summary
df.shape
df.info()
df.describe()
```

```python
# Data Handling and Cleaning
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Univariate Analysis
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Income'], kde=True)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Family', data=df)
```

```python
# Bivariate Analysis
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Experience', y='Income', data=df)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Family', y='Income', data=df)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Education', y='Income', data=df)
```

```python
# Multivariate Analysis
# Selecting only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include='number')
# Heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.pairplot(df[['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']])
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Education', y='Income', hue='Personal Loan', data=df)
```
